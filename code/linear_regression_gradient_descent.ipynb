```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt



# Generate synthetic dataset
np.random.seed(42)  # For reproducibility
X = 2 * np.random.rand(100, 1)  # Random input feature
y = 4 + 3 * X + np.random.randn(100, 1)  # Linear relationship with noise

# Add bias term (intercept) to input features
X_b = np.c_[np.ones((100, 1)), X]  # Bias term (x0 = 1)

# Initialize parameters
theta = np.random.randn(2, 1)  # Random initial weights (intercept and slope)
learning_rate = 0.1  # Step size for updates
iterations = 100  # Number of iterations for Gradient Descent
m = len(X_b)  # Number of training examples

# Lists to track cost function and parameter updates
cost_history = []
theta_history = []

# Gradient Descent loop
for _ in range(iterations):
    gradients = 2 / m * X_b.T.dot(X_b.dot(theta) - y)  # Compute gradients
    theta -= learning_rate * gradients  # Update parameters
    cost = np.mean((X_b.dot(theta) - y) ** 2)  # Compute cost (MSE)
    cost_history.append(cost)  # Store cost function value
    theta_history.append(theta.copy())  # Store parameter updates

# Final parameters after optimization
print("Optimized Parameters (Î¸):", theta)

# Visualization 1: Cost Function Convergence
plt.figure(figsize=(10, 6))
plt.plot(range(iterations), cost_history, linestyle='--', marker='o', label="Cost Function")
plt.title("Cost Function Convergence Over Iterations")
plt.xlabel("Iterations")
plt.ylabel("Mean Squared Error")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Visualization 2: Parameter Update Trajectories
theta_history = np.array(theta_history).reshape(-1, 2)  # Reshape for plotting
plt.figure(figsize=(10, 6))
plt.scatter(theta_history[:, 0], theta_history[:, 1], c=range(len(theta_history)), cmap='viridis', label="Parameter Updates")
plt.colorbar(label="Iterations")
plt.scatter(theta[0], theta[1], color='red', s=100, label="Optimal Parameters")
plt.title("Parameter Updates Trajectory")
plt.xlabel("Î¸0 (Intercept)")
plt.ylabel("Î¸1 (Slope)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
